```python
#Heatmap for OT data
sns.heatmap(qb_data.corr())
qb_var = {}
for col in qb_data:
    if (qb_data.var().get(col) != None):
        qb_var[col] = (qb_data.var().get(col))
```

```python
lists = sorted(qb_var.items()) # sorted by key, return a list of tuples

x, y = zip(*lists) # unpack a list of pairs into two tuples

plt.plot(x, y, linewidth=3)
plt.xticks(fontsize=14, rotation=90)
plt.tight_layout()
plt.show()
```
